```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.txt')
df.columns = ['sudoku','rec','bt1','bt2']
df.info()
```

```python
df2 = df.groupby(df.sudoku).mean()
df2.info()
```

```python
sns.histplot(data=df2, x="rec",color="blue", bins = 2000, label="Recursive")
sns.histplot(data=df2, x="bt1",color="green", bins = 2000, label="Slow Backtrack")
sns.histplot(data=df2, x="bt2",color="red", bins = 2000, label="Quick Backtrack")
plt.yscale('log')
plt.ylim([1, 10000])
plt.xlim([0, 15])
plt.legend()
plt.xlabel('Solve Time [ms]')
```

```python
sns.histplot(data=df2, x="rec",color="blue", bins = 10000, label="Recursive")
sns.histplot(data=df2, x="bt1",color="green", bins = 10000, label="Slow Backtrack")
sns.histplot(data=df2, x="bt2",color="red", bins = 10000, label="Quick Backtrack")
plt.yscale('log')
plt.ylim([1, 100000])
plt.xlim([0, 3])
plt.legend()
plt.xlabel('Solve Time [ms]')
```
